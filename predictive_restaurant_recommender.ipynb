### feature engineering

```python

import numpy as np
```

```python
## calculating distance from customer to vendor and adding it as column



def haversine(lat1,lon1,lat2,lon2):
    R=6371

    lat1,lon1,lat2,lon2=map(np.radians,[lat1,lon1,lat2,lon2])
    dlat=lat2-lat1
    dlon=lon2-lon1

    a=np.sin(dlat/2)**2 + np.cos(lat1)*np.cos(lat2)*np.sin(dlon/2)**2

    c=2*np.arcsin(np.sqrt(a))

    return R*c

master_df["distance_km"]=haversine(master_df["customer_latitude"],master_df["customer_longitude"],master_df["vendor_latitude"],master_df["vendor_longitude"])

```

Creating customer-vendor pairs

Each row currently represents only the chosen vendor.
To enable comparison, we generate additional rows with other vendors as negative samples.

We convert each row into multiple rows:
* 1 positive (chosen vendor)
* multiple negative (not chosen vendors)

This helps the model learn how customers compare vendors

```python
##  dropping data leakage columns

base_df=master_df_new.drop(columns=["customer_id","deliverydistance","preparationtime","prepration_time"])
base_df
```

```python
base_df.columns
```

```python
customer_df=base_df.drop(columns=["vendor_category_en_sweets_and_bakes",
                                  "vendor_category_en_restaurants",
                                  "delivery_charge",
                                  "serving_distance",
                                  "is_open",
                                  "vendor_rating"])

customer_df
```

```python
customer_df.columns
```

```python
vendor_df = base_df[
    [
        'vendor_id',
        'vendor_rating',
        'delivery_charge',
        'serving_distance',
        'is_open',
        'vendor_category_en_restaurants',
        'vendor_category_en_sweets_and_bakes'
    ]
].drop_duplicates('vendor_id')
```

```python
vendor_df
```

```python
vendor_df.columns
```

```python
## get all unique vendors

all_vendors=vendor_df["vendor_id"].unique()
all_vendors
```

```python
## pair creation

import random

rows=[]

for i,row in customer_df.iterrows():

    chosen_vendor=row["vendor_id"]

    # positive
    postve_rows=row.copy()
    postve_rows["Label"]=1
    rows.append(postve_rows)


    # negative

    negtve_vendors=random.sample(list(all_vendors),2)

    for v in negtve_vendors:
        if v==chosen_vendor:
            continue

        neg_row=row.copy()
        neg_row["vendor_id"]=v
        neg_row["Label"]=0
        rows.append(neg_row)




```

```python
rows
```

```python
pairs_df=pd.DataFrame(rows)
pairs_df
```

```python
pairs_df.shape
```

```python
pairs_df["Label"].value_counts()
```

```python
pairs_df.head()
```

```python
final_df=pairs_df.merge(vendor_df,on="vendor_id",how="left")
final_df
```

```python
final_df.shape
```

```python
final_df.columns
```

```python
X=final_df.drop(columns=["vendor_id","Label"])
y=final_df["Label"]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=65)

```

```python
from sklearn.ensemble import RandomForestClassifier

model=RandomForestClassifier(random_state=43)

model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy=accuracy_score(y_test,y_pred)

print("accuracy",accuracy)
```

```python
from sklearn.metrics import classification_report,roc_auc_score

print(classification_report(y_test,y_pred))

y_prob=model.predict_proba(X_test)[:,1]

print("ROC-AUC",roc_auc_score(y_test,y_prob))
```
